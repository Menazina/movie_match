Vamos a usar sistemas de recomendación. Basado en distancia.

```python
distance = squareform(pdist(genre.iloc[:,:-3], 'euclidean'))
```

```python
distance.shape
```

```python
similarity = 1/(1+distance)
```

```python
similarities = pd.DataFrame(similarity, index = genre.index, columns = genre.index)
```

```python
similarities
```

```python
similarities.columns = genre["Title"]
```

```python
similarities.index = genre["Title"]
```

```python
similarities
```

```python
def plot(m1, m2):
    
    x=similarities.T[m1]
    y=similarities.T[m2]
    
    n=list(similarities.T.index)
    
    plt.figure(figsize=(10, 5))
    
    plt.scatter(x, y, s=0)
    
    plt.title('Distance between {} VS. {}'.format(m1, m2), fontsize=14)
    plt.xlabel(m1, fontsize=14)
    plt.ylabel(m2, fontsize=14)
    
    for i,e in enumerate(n):
        plt.annotate(e, (x[i], y[i]), fontsize=12)
    plt.show();
```

```python
plot("Prometheus","Sing")
```
